```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
```

```python
type(dataset)
```

```python
dataset
```

```python
x=dataset.iloc[:,1:4].values
```

```python
x
```

```python
y=dataset.iloc[:,-1:].values
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
x[:,0]=lb.fit_transform(x[:,0])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0) 
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```

```python
import sklearn.metrics as metrics
```

```python
fpr,tpr, threshlod=metrics.roc_curve(y_test,y_predict)
```

```python
roc_auc=metrics.auc(fpr,tpr)
```

```python
roc_auc
```

```python
plt.title("Receiver Operating Characteristics")
plt.plot(fpr,tpr,'b',label='AUC=%0.2f'%roc_auc,color='blue')
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier_knn.fit(x_train, y_train)
```

```python
y_pred_knn = classifier_knn.predict(x_test)
```

```python
import sklearn.metrics as metrics
fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred_knn)
roc_auc_knn= metrics.auc(fpr, tpr)
```

```python
from sklearn.svm import SVC
classifier_svc = SVC(kernel = 'linear')
classifier_svc.fit(x_train, y_train)
```

```python
y_pred_svc = classifier_svc.predict(x_test)
```

```python
import sklearn.metrics as metrics
fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred_svc)
roc_auc_svc = metrics.auc(fpr, tpr)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier_dt = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier_dt.fit(x_train, y_train)
```

```python
y_pred_dt = classifier_dt.predict(x_test)
```

```python
import sklearn.metrics as metrics
fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred_dt)
roc_auc_dt = metrics.auc(fpr, tpr)
```

```python
x=["LR","KNN","Linear SVM","DT"]
```

```python
data=[roc_auc,roc_auc_knn,roc_auc_svc,roc_auc_dt]
```

```python
import matplotlib.pyplot as plt
plt.title('AUC')
plt.bar(x, data)
plt.show()
```
